```python
import os
import requests
import pandas as pd
import helper as hlp
from request_wrapper import request_wrapper
```

Get assets offered by cointec for exchange against quote currency GBP:

```python
gbp_assets = request_wrapper(hlp.get_assets_cointec,1).json()
gbp_assets = [key[3:len(key)] for key, value in gbp_assets.items() if 'GBP' in key]

df_tickers = pd.DataFrame(gbp_assets)
```

```python
df_tickers.iloc[:15]
```

Iterate through cointec assets and check if historical chart data is available for each asset on **Bitcoin Average**.

```python
status_btc_avg = []

for i in gbp_assets:
    response = request_wrapper(hlp.get_charts_btc_average,5,i)
    
    if response.status_code == 200:
        status_btc_avg.append(True)
        
    elif response.status_code == 400:
        status_btc_avg.append(False)
        
    else:
        print response
        break
        
print status_btc_avg
```

Iterate through cointec assets and check if historical chart data is available for each asset on **Coin API**

```python
status_coinapi =[]

for i in gbp_assets:
    response = request_wrapper(hlp.get_charts_coinapi,5,i)    
    if response.status_code == 200:
        
        if len(response.json()) != 0:
            status_coinapi.append(True)
        
        elif len(response.json()) == 0:
            status_coinapi.append(False)
        
    else:
        print response.status_code,response.text
        break
        
print status_coinapi
```

Iterate through cointec assets and check if historical chart data is available for each asset on **Crypto Compare**

```python
status_cryptocompare = []

for i in gbp_assets:
    response = request_wrapper(hlp.get_charts_cryptocompare,5,i)    
    if response.status_code == 200:
        
        if response.json()["Response"] == 'Success':
            status_cryptocompare.append(True)
        
        elif response.json()["Response"] == 'Error':
            status_cryptocompare.append(False)
        
    else:
        print response.status_code,response.text
        break
        
print status_cryptocompare
```

Iterate through cointec assets and check if historical chart data is available for each asset on **Poloniex**

```python
status_poloniex = []

for i in gbp_assets:
    response = request_wrapper(hlp.get_charts_poloniex,5,i)
    if response.status_code == 200:
        
        if 'error' not in response.json():
            status_poloniex.append(True)
            
        elif 'error' in response.json():
            status_poloniex.append(False)
            
    else:
        print response.status_code,response.text
        break
        
print status_poloniex
```

```python
df_charts = pd.DataFrame(data={'btc avg':status_btc_avg,'coin api':status_coinapi,
                         'cryptocompare':status_cryptocompare,'poloniex':status_poloniex},
                         index = gbp_assets
                        )
```
